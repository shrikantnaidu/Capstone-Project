```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import datetime
import time
import swifter
```

```python
final_data_1 = pd.read_csv('./final_data_1.csv')
final_data_1.head()
```

```python
final_data_2 = pd.read_csv('./final_data_2.csv')
final_data_2.head()
```

```python
print("Shape 1:",final_data_1.shape)
print("Shape 2:",final_data_2.shape)
```

```python
final_data = pd.DataFrame()
final_data = final_data.append(final_data_1)
final_data.shape
```

```python
final_data = final_data.append(final_data_2)
final_data.shape
```

```python
final_data.head()
```

```python
final_data = final_data[['UserId','Category', 'Country',  'Device',  'Genre', 'OS', 'ProgramType', 'State', 'VideoId','Event', 'EventFrequency','DateTime']]
final_data.head()
```

```python
final_data.sort_values(['UserId','DateTime'],axis=0,inplace=True)
final_data.head()
```

```python
final_data['PreviousDateTime'] = final_data['DateTime'].copy()
final_data.head()
```

```python
final_data['PreviousDateTime'] = final_data.groupby(['UserId'])['PreviousDateTime'].shift(1)
```

```python
final_data.head()
```

```python
final_data.isnull().sum()
```

# Filling NaN values for subset 1

```python
final_data_one = final_data[(final_data['EventFrequency'] > 1)]
final_data_one.head()
```

```python
final_data_one.isnull().sum()
```

```python
final_data_one['PreviousDateTime'] =  final_data_one['PreviousDateTime'].fillna(method='bfill',axis=0)
final_data_one.head()
```

# Filling NaN values for subset 2

```python
final_data_two = final_data[final_data['EventFrequency'] == 1]
final_data_two.head()
```

```python
final_data_two.isnull().sum()
```

```python
print("Subset1:",final_data_one.shape)
print('Subset2:',final_data_two.shape)
print("Main:",final_data.shape)
```

```python
final_data_two['PreviousDateTime'] = final_data_two['PreviousDateTime'].where((final_data_two['PreviousDateTime'].notnull()),final_data_two['DateTime'])
```

```python
final_data_two.isnull().sum()
```

# Merging the two subsets

```python
data = pd.DataFrame()
data = data.append(final_data_one)
data = data.append(final_data_two)
```

```python
print("New:",data.shape)
print("Main:",final_data.shape)
```

```python
data.isnull().sum()
```

```python
data.sort_values(['UserId','DateTime'],axis=0,inplace=True)
```

```python
data.to_csv(path_or_buf='merged_data.csv',sep=',',index=False)
```
