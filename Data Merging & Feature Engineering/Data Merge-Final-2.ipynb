```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import datetime
import time
import swifter
```

# Data Loading

# Video Started

```python
VideoStartedMod=pd.read_csv('./data_modded/VideoStartedMod.csv',sep=',')
VideoStartedMod.head()
```

```python
VideoStartedMod.sort_values(['UserId','DateTime'],axis=0,inplace=True)
VideoStartedMod.head()
```

```python
VideoStartedMod['Event'] = "Video Started"
user_frequency = VideoStartedMod['UserId'].value_counts()
VideoStartedMod = VideoStartedMod.merge(user_frequency,how='inner',left_on='UserId',right_index=True)
VideoStartedMod.drop(['UserId_x'],axis=1,inplace=True)
VideoStartedMod.rename({'UserId_y':'EventFrequency'},axis='columns',inplace=True)
VideoStartedMod.head()
```

```python
VideoStartedMod = VideoStartedMod[['UserId', 'State', 'Genre', 'Category', 'ProgramType', 'Country', 'Device', 'OS', 'VideoId',  'Event', 'EventFrequency','DateTime']]
VideoStartedMod.head()
```

# Video Details

```python
VideoDetailsMod=pd.read_csv('./data_modded/VideoDetailsMod.csv',sep=',')
VideoDetailsMod.head()
```

```python
VideoDetailsMod.sort_values(['UserId','DateTime'],axis=0,inplace=True)
VideoDetailsMod.head()
```

```python
VideoDetailsMod['Event'] = "Video Details"
user_frequency = VideoDetailsMod['UserId'].value_counts()
VideoDetailsMod = VideoDetailsMod.merge(user_frequency,how='inner',left_on='UserId',right_index=True)
VideoDetailsMod.drop(['UserId_x'],axis=1,inplace=True)
VideoDetailsMod.rename({'UserId_y':'EventFrequency'},axis='columns',inplace=True)
VideoDetailsMod.head()
```

```python
VideoDetailsMod = VideoDetailsMod[['UserId', 'State', 'Genre', 'Category', 'ProgramType', 'Country', 'Device', 'OS', 'VideoId',  'Event', 'EventFrequency','DateTime']]
VideoDetailsMod.head()
```
