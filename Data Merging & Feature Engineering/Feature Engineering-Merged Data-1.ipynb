```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import datetime
import time
import swifter
```

```python
data = pd.read_csv('./merged_data.csv')
```

```python
data.head(20)
```

```python
data.isnull().sum()
```

```python
data['LastSeenTimestamp'] = data['DateTime'].astype('datetime64[ns]')  - data['PreviousDateTime'].astype('datetime64[ns]') 
data.head()
```

```python
def diffHours(dd) :
    days, seconds = dd.days, dd.seconds
    hours = days * 24 + seconds // 3600
    return hours

data['LastSeen(in Hours)'] = data['LastSeenTimestamp'].swifter.apply(lambda x : diffHours(x))
data.head()
```

```python
data['LastSeen(in Hours)'].value_counts()
```

```python
data['LastSeen(in Hours)'] = data['LastSeen(in Hours)'].where(data['LastSeen(in Hours)'] > 0,0)
```

```python
data['LastSeen(in Hours)'].value_counts()
```

```python
data.head()
```

# Tag an event if within 48 hours

```python
data['LastEvent(in 48H)'] = data['LastSeen(in Hours)'].swifter.apply(lambda x :'Yes' if x <= 48.0 else 'No')
```

```python
data['WatchedVideo'] = 1
data['WatchedVideo'] = data['WatchedVideo'].where((data['Event']!='Video Started'),0)
```
